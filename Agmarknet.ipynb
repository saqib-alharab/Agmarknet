__This Doesnt Give Much Insight We Need to Seperate each to gain More Insights__

```python
#Seperating Markets For Price Pattern Analysis

Achnera = data[data["Market Name"]=='Achnera']
Fatehpur_Sikri =  data[data["Market Name"]=='Fatehpur Sikri']
Agra = data[data["Market Name"]=='Agra']
Fatehabad = data[data["Market Name"]=='Fatehabad']
Jagnair = data[data["Market Name"]=='Jagnair']
Khairagarh =data[data["Market Name"]=='Khairagarh']
Samsabad = data[data["Market Name"]=='Samsabad']
Jarar = data[data["Market Name"]=='Jarar']
```

```python
#Visualisisng Price Pattern Across Different Markets 

plt.style.use('ggplot')
plt.figure(figsize=(30,20))
plt.subplot(241)
plt.title('Achnera')
sns.lineplot(data=Achnera, x="Price Date", y="Modal Price (Rs./Quintal)")
plt.subplot(242)
plt.title('Agra')
sns.lineplot(data=Agra, x="Price Date", y="Modal Price (Rs./Quintal)")
plt.subplot(243)
plt.title('Fatehpur_Sikri')
sns.lineplot(data=Fatehpur_Sikri, x="Price Date", y="Modal Price (Rs./Quintal)")
plt.subplot(244)
plt.title('Fatehabad')
sns.lineplot(data=Fatehabad, x="Price Date", y="Modal Price (Rs./Quintal)")
plt.subplot(245)
plt.title('Jaganair')
sns.lineplot(data=Jagnair, x="Price Date", y="Modal Price (Rs./Quintal)")
plt.subplot(246)
plt.title('Khairagarh')
sns.lineplot(data=Khairagarh, x="Price Date", y="Modal Price (Rs./Quintal)")
plt.subplot(247)
plt.title('Samsabad')
sns.lineplot(data=Samsabad, x="Price Date", y="Modal Price (Rs./Quintal)")
plt.subplot(248)
plt.title('Jarar')
sns.lineplot(data=Jarar, x="Price Date", y="Modal Price (Rs./Quintal)")
plt.show()
```

__There is an interesting pattern in almost all the Markets as in first 2months of 2020 the price daily reduced and from 3rd month it started increasing exponentially. Till 7th month the increase was mild numbers but after 7th it was drastic eventually the price was at peak in the last two months of 2020__

__The cause of the Price Hike is due to lockdown heavy usage and low transportation reported by some traders and by November cultivation wasn’t matching the supply that’s why it was at peak__
